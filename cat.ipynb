```python
import pandas as pd

# Charger ton CSV
csv_file = 'df_final_normalise_fr.csv'
df = pd.read_csv(csv_file)

# V√©rifier que la colonne existe
if 'Category' not in df.columns:
    print("‚ùå ERREUR : La colonne 'Category' est absente du CSV.")
    exit()

# Nettoyage l√©ger (s√©curit√©)
df['Category'] = df['Category'].astype(str).str.strip()

# Nombre total de cat√©gories uniques
unique_categories = sorted(df['Category'].unique())
print(f"‚úÖ Nombre de cat√©gories uniques : {len(unique_categories)}\n")

# Affichage des cat√©gories tri√©es
print("üìã Liste des cat√©gories :")
for cat in unique_categories:
    print("  -", cat)

# V√©rification de doublons possibles (m√™me nom mais majuscules ou espaces)
normalized = df['Category'].str.lower().str.strip()
if len(normalized.unique()) != len(df['Category'].unique()):
    print("\n‚ö†Ô∏è Attention : certaines cat√©gories semblent identiques √† cause de majuscules ou d'espaces !")
    print("Par exemple :", set(df['Category']) - set(normalized))
else:
    print("\n‚úÖ Aucune incoh√©rence d√©tect√©e dans les cat√©gories.")
```
